```python
import pandas as pd
import requests
import time, urllib3
urllib3.disable_warnings()
```

https://data.oncb.go.th/casePer

```python
def get_data(year):
    response = requests.get(f"https://dataapi.oncb.go.th/suppress/case_per/{year}", verify = False)
    if response.ok:
        return pd.DataFrame(response.json()["data"])
    return None
```

```python
df = get_data(2557)
for year in range(2558, 2566):
    df_append = get_data(year)
    df = pd.concat([df, df_append]).copy()
    time.sleep(.5)

df = df.reset_index(drop=True).copy()
df.head()
```

```python
df.to_csv("../data/arrested_drug_cases.csv")
```
